# Exploratory Data Analysis

# Gender

```python
df.columns
```

```python
ax=sns.countplot(x="Gender",data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_gen=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sales_gen
```

```python
sales_gen=df.groupby(['Gender'],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
sns.barplot(x='Gender',y='Amount',data=sales_gen)
```

```python
From above graph we can see that most of t buyers are females and even the purchasing power of females are grater than men.
```

# Age

```python
ax= sns.countplot(data=df, x= "Age Group", hue= 'Gender')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Total amount vs Age group
sales_age=df.groupby(["Age Group"], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(data= sales_age,x='Age Group',y= 'Amount')
```

```python
From above graph we can see that most of the buyers are of age group 26-35 yrs female
```

# State

```python
# total number of orders from top 10 states
sales_state=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.set(rc={'figure.figsize':(40,10)})
sns.barplot(data=sales_state,x='State',y='Orders')
```

```python
# total amount/sales from top 10 states
sales_state= df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sales_state['Amount']=sales_state['Amount'].round(2)
sns.set(rc={'figure.figsize':(40,10)})
ax=sns.barplot(data=sales_state,x='State',y='Amount')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
From above graphs we can see that most of the orders and total sales/amount are from  Uttar Pradesh, Maharashtra and Karnataka respectively.
```

# Marital Status

```python
ax=sns.countplot(data=df,x='Marital_Status')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales = df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.set(rc={'figure.figsize':(5,5)})
sns.barplot(data=sales, x='Marital_Status', y='Amount', hue='Gender')
```

```python
From above grapgh we can see that most of the buyers are married(women) and they have high purchasing power.
```

# Occupation

```python
sns.set(rc={'figure.figsize':(40,10)})
ax=sns.countplot(x='Occupation',data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales=df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
plt.figure(figsize=(20,5))
sns.lineplot(data=sales,x='Occupation',y='Amount',markers='0',label='Total Amount by Occupation')
plt.xlabel('occupation')
plt.ylabel('Amount')
plt.title('Total Amount by Occupation')
plt.legend()
plt.show()
```

```python
# from above graph we can see that most of the buyers are from IT, Healthcare and Aviation sector.
```

# Product Category

```python
sns.set(rc={'figure.figsize':(40,10)})
ax=sns.countplot(x='Product_Category',data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales=df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,10)})
ax=sns.barplot(data=sales,x='Product_Category',y='Amount')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# From above graph we can see that most of the sold products are from food,clothing and electronic category.
```

```python
sales=df.groupby(['Product_ID'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending= False).head(10)
sns.set(rc={'figure.figsize':(40,10)})
ax=sns.barplot(data=sales,x='Product_ID',y='Orders')
for bars in ax.containers:
    ax.bar_label(bars)
```

# Conclusion

```python
Married women age group 26-35 yrs from UP, Maharashtra and Karnataka working in IT, Healthcare and Aviation are more likely 
to buy products from food, clothing and electronics category.
```
